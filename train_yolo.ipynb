```python
import torch
from ultralytics import YOLO

def main():
    # 1. Verify your system has access to hardware acceleration (GPU)
    device = "cuda" if torch.cuda.is_available() else "cpu"
    print(f"Training will execute on: {device.upper()}")

    # 2. Load the edge-optimized nano weights (automatically downloads weights)
    # yolo26n sits at the cutting edge for local/CPU execution speed
    model = YOLO("yolo26n.pt")

    # 3. Kick off the training configuration
    model.train(
        data="./FOOD-INGREDIENTS-dataset-4/data.yaml", # Path to your config roadmap
        epochs=100,                        # 100 epochs is plenty for a clear signal on 4-10 classes
        patience=30,
        imgsz=640,                         # Resizes all inputs to standard mobile layout
        batch=-1,                          # Lower to 16 if your GPU hits "Out of Memory" errors
        workers=4,                         # Number of CPU threads loading data concurrently
        device=device,                     # Directs processing to the GPU
        plots=True,                         # Generates loss charts and PR curves automatically
        augment=True,
        mosaic=1.0,            # composites multiple images — helps multi-ingredient scenes
        flipud=0.3,            # vertical flip — ingredients photographed from above
        degrees=15,            # rotation — fridge/counter angles vary
        hsv_h=0.015,           # slight hue shift — handles different lighting
        hsv_s=0.5,             # saturation variance
        hsv_v=0.4,             # brightness variance — fridge lighting is dim
        project="food_detector",
        name="yolo26s_ingredients",
    )

if __name__ == "__main__":
    main()
```
